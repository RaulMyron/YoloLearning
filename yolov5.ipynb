```python
import torch
from matplotlib import pyplot as plt
from PIL import Image, ImageDraw, ImageFont
import numpy as np
import os
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
```

```python
def resize_image(img, base_width):

    w_percent = (base_width / float(img.size[0]))
    h_size = int((float(img.size[1]) * float(w_percent)))
    return img.resize((base_width, h_size), Image.LANCZOS)
    
img_path = 'src/kitty.jpg'

# Load the image
img = Image.open(img_path)

# Resize the image while keeping aspect ratio
base_width = 300
img_resized = resize_image(img, base_width)

# Perform detection
results = model(img_resized)

# Display results
results.show()

# Save the image with detections
results.save(save_dir='path/to/save')

# Plot the image with detections in Jupyter Lab
plt.figure(figsize=(6, 4))
plt.imshow(np.array(img_resized))
plt.axis('off')
plt.show()
```

```python
img_dir = 'src'
save_dir = 'src/yolov5/'

# Ensure save directory exists
os.makedirs(save_dir, exist_ok=True)

# Process each image in the directory
for img_file in os.listdir(img_dir):
    if img_file.endswith(('.jpg', '.jpeg', '.png', '.bmp', '.gif')):
        img_path = os.path.join(img_dir, img_file)

        # Load the image
        img = Image.open(img_path)

        # Resize the image while keeping aspect ratio
        base_width = 700
        img_resized = resize_image(img, base_width)

        # Perform detection
        results = model(img_resized)

        # Convert the results to a format suitable for visualization
        detections = results.xyxy[0].numpy()  # [x1, y1, x2, y2, confidence, class]

        # Draw the bounding boxes and labels
        draw = ImageDraw.Draw(img_resized)
        font = ImageFont.load_default()

        for detection in detections:
            x1, y1, x2, y2, confidence, cls = detection
            label = f"{model.names[int(cls)]} {confidence:.2f}"
            draw.rectangle([x1, y1, x2, y2], outline="red", width=2)
            draw.text((x1, y1), label, fill="red", font=font)

        # Save the image with detections
        save_path = os.path.join(save_dir, img_file)
        img_resized.save(save_path)

        # Plot the image with detections in Jupyter Lab
        plt.figure(figsize=(6, 4))
        plt.imshow(np.array(img_resized))
        plt.axis('off')
        plt.show()
```
